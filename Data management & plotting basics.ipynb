# Data Management

Data management is key in biology. Here we will use the `pandas` package.

In order to be able to exploit it correctly the data needs to be well structured. The easiest way to make sure the data will be usable is to make it __tidy__ : one line = one observation, one column = one variable.



```python
import pandas as pd
```

## Basic Scatter plot

The syntax is really easy when combined with a dataframe (what `pandas` manages).
* x = the column you want on the x axis
* y = the column you want on the y axis
* data = the data you want to use
* hue = the column you want to use to group the data, the different classes will be displayed as a different color 
* style = idem as hue but with a different shape (can be used simultaneously) 

```python
sns.scatterplot(x="petal_length",y="sepal_length",data=df,hue="class")
```

## Automation
A quick exemple of automation that can be done in python to efficiently generate figures.

```python
plt.figure(figsize=(8,8))
list_variables=["sepal_length","petal_length","sepal_width","petal_width"]
for i in range(4):
    plt.subplot(2,2,i+1)
    sns.violinplot(x="class",y=list_variables[i],data=df,palette="viridis")
    plt.ylim(0,8)
plt.suptitle("All Variables & All Species")
```

```python
import scipy.stats as sst
```

Let's say we want to do a t-test between the sepal length of two of the specices.

```python
df_versicolor=df.loc[df['class']=="Iris-versicolor"]
df_setosa=df.loc[df['class']=="Iris-setosa"]
```

We first identify our two groups.

```python
sst.ttest_ind(df_versicolor.sepal_length,df_setosa.sepal_length)
```
